## Dimensionality reduction

It is very important if two two input values are really close in high dimensinal space then they should be definitlt be close in low dimensional space. 


So the question is that how to do this? We need to find a function. In NN we introduce the cost function and we want to minimize it. Here we want to defince something that we can minimize it. 


$x_i-x_j$ is the distance in higher dimension and $y_i-y_j$ is the distance in lower dimension.   

The idea of t-SNE is that when picking pairs of points $p_i$ and set up a probability distribution that prefers points that are close to each other. To choose this randomly a pair of points will become more likely if they are close to each other. $p_{ij}$ is the probability to pick of a pair of points and it will be a large if two points are close. $q_{ij}$ is similarly a probability distribution but in the low dimensional space. It will be large if the points are close. These probablities are normalizd : $\sum_{i\neq j} q_{ij} = 1$, $\sum_{i\neq j} p_{ij} = 1$.    

These two probabilities shoud be similar to each other as much as possible. This is kine of distance measure to chack the similarity of these probabilities, it is called Kullback-Leibler divergence.
